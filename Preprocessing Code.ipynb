```python
import os
import glob
import shutil
from PIL import Image
from sklearn.model_selection import train_test_split


data_dir = "C:/Users/vaishnavi.gejji/Downloads/CapstoneProject/Brain Tumor Data Set"
```

```python
print(os.listdir(data_dir))
```

```python

#=========================================
# 🔹 3. REMOVE CORRUPTED IMAGES
# =========================================
def remove_corrupted_images(folder):
    for root, _, files in os.walk(folder):
        for file in files:
            path = os.path.join(root, file)
            try:
                img = Image.open(path)
                img.verify()
            except:
                os.remove(path)
                print("Removed corrupted:", path)

remove_corrupted_images(data_dir)

print("✅ Cleaning done")
```

```python

# =========================================
# 🔹 4. READ IMAGE PATHS
# =========================================
image_paths = glob.glob(data_dir + "/**/*.jpg", recursive=True) + \
              glob.glob(data_dir + "/**/*.png", recursive=True) + \
              glob.glob(data_dir + "/**/*.jpeg", recursive=True)

print("Total images:", len(image_paths))
```

```python

# =========================================
# 🔹 5. CREATE LABELS
# =========================================
def get_label(path):
    if "Healthy" in path:
        return 0
    else:
        return 1

labels = [get_label(p) for p in image_paths]
```

```python

# =========================================
# 🔹 6. TRAIN / VAL / TEST SPLIT
# =========================================
train, temp, y_train, y_temp = train_test_split(
    image_paths, labels, test_size=0.3, random_state=42
)

val, test, y_val, y_test = train_test_split(
    temp, y_temp, test_size=0.5, random_state=42
)

print("Train:", len(train))
print("Validation:", len(val))
print("Test:", len(test))
```

```python

# =========================================
# 🔹 7. CREATE PROCESSED FOLDER
# =========================================
base_dir = "processed_data"

train_dir = os.path.join(base_dir, "train")
val_dir   = os.path.join(base_dir, "val")
test_dir  = os.path.join(base_dir, "test")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python

# =========================================
# 🔹 8. PREPROCESS + SAVE IMAGES
# =========================================
def process_and_save(file_list, target_dir):
    for file in file_list:
        try:
            img = Image.open(file).convert("RGB")
            img = img.resize((224, 224))
            filename = os.path.basename(file)
            img.save(os.path.join(target_dir, filename))
        except:
            continue

process_and_save(train, train_dir)
process_and_save(val, val_dir)
process_and_save(test, test_dir)

print("✅ Preprocessing + saving done")

```

```python

# =========================================
# 🔥 🔹 9. CREATE ZIP FILE (LOCAL)
# =========================================
shutil.make_archive("processed_data", 'zip', base_dir)

print("✅ ZIP created: processed_data.zip")
```
